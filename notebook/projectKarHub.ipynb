Join - das duas tabelas

```python
from google.cloud import bigquery
from google.oauth2 import service_account

def process_data(df, bid_rate, tipo):
    df[['ID Fonte Recurso', 'Nome Fonte Recurso']] = df['Fonte de Recursos'].str.extract(r'^(\d+) - (.*)$')

    df[tipo] = df[tipo].str.replace('.', '').str.replace(',', '.').astype(float)

    grouped_df = df.groupby(['ID Fonte Recurso', 'Nome Fonte Recurso']).agg({
        tipo: 'sum'
    }).reset_index()

    grouped_df[tipo] = grouped_df[tipo] * bid_rate

    return grouped_df

def data_join(grouped_df_receita, grouped_df_despesas):

    df = pd.merge(grouped_df_receita, grouped_df_despesas, on=['ID Fonte Recurso', 'Nome Fonte Recurso'], how='outer')
    df['Arrecadado'].fillna(0.00, inplace=True)
    df['Liquidado'].fillna(0.00, inplace=True)
    df['dt_insert'] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    df['dt_insert'] = pd.to_datetime(df['dt_insert'])

    return df

url = "https://economia.awesomeapi.com.br/json/daily/USD-BRL?start_date=20220622&end_date=20220622"

def request_api_dolar(url):
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        if len(data) > 0:
            return float(data[0]['bid'])
```

```python
bid_rate = request_api_dolar(url)

grouped_df_receita = process_data(df_receita, bid_rate, 'Arrecadado')
grouped_df_despesas = process_data(df_despesas, bid_rate, 'Liquidado')

df_merged = data_join(grouped_df_receita, grouped_df_despesas)
```

```python
df_merged = pd.merge(grouped_df_receita, grouped_df_despesas, on=['ID Fonte Recurso', 'Nome Fonte Recurso'], how='outer')
df_merged['Arrecadado'].fillna(0.00, inplace=True)
df_merged['Liquidado'].fillna(0.00, inplace=True)
df_merged['Arrecadado'] = df_merged['Arrecadado'] * bid_rate
df_merged['Liquidado'] = df_merged['Liquidado'] * bid_rate
df_merged["dt_insert"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
df_merged['dt_insert'] = pd.to_datetime(df_merged['dt_insert'])
```

```python
df_merged
```

```python
df_merged['Margem Bruta'] = df_merged['Arrecadado'] - df_merged['Liquidado']
df_margem_bruta = df_merged.groupby(['ID Fonte Recurso', 'Nome Fonte Recurso'])['Margem Bruta'].sum().reset_index()
df_margem_bruta = df_margem_bruta.sort_values(by='Margem Bruta', ascending=False)
top_5_fontes = df_margem_bruta.head(5)
print("As 5 fontes de recursos com melhor margem bruta são:")
print(top_5_fontes[['ID Fonte Recurso', 'Nome Fonte Recurso', 'Margem Bruta']])
```
